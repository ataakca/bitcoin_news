# Tokenization

```python
# I couldn't figure out a way to filter the abbreviations, therefore I replaced the U.S.
# to United-States.
for i in range(len(all_news['article'])):
    all_news.loc[:,'article'][i] = all_news.loc[:,'article'][i].replace('U.S.', 'United States')
    
for x in range(len(data['article'])):
    data.loc[:,'article'][x] = data.loc[:,'article'][x].replace('U.S.', 'United States')
```

```python
# Tokenizing both of our datasets
tokenizer = RegexpTokenizer(r"[\w+']+[\w+-]+")
tokens = []
for article in range(len(all_news['article'])):
    tokens.append(tokenizer.tokenize(all_news['article'][article]))
    
data_tokens = []
for article in range(len(data['article'])):
    data_tokens.append(tokenizer.tokenize(data['article'][article]))
```

```python
# To create a list that has identical number of lists inside it.
# This is done to prevent overwriting the tokens
lemmatized_tokens = tokens
```

```python
# # Tagging the words to lemmatize them according to their contextual meaning (verb, noun, adj)
# # source : https://www.digitalocean.com/community/tutorials/how-to-perform-sentiment-analysis-in-python-3-using-the-natural-language-toolkit-nltk
# # I had to make some small changes in order to keep the tokens belonged to 
# # different articles separated to not lose the context.
# def lemmatize_sentence(tokens_to_be_lemmatized):
#     lemmatizer = WordNetLemmatizer()
#     lemmatized_sentence = []
#     for i in range(len(lemmatized_tokens)):
#         lemmatized_sentence=[]
#         for word, tag in pos_tag(lemmatized_tokens[i]):
#             if tag.startswith('NN'):
#                 pos = 'n'
#             elif tag.startswith('VB'):
#                 pos = 'v'
#             else:
#                 pos = 'a'
#             lemmatized_sentence.append(lemmatizer.lemmatize(word, pos))
#             lemmatized_tokens[i] = lemmatized_sentence
#     return lemmatized_sentence
```

```python
# # Changing the lemmatized tokens list into actual lemmatized tokens
# lemmatized_sentences = lemmatize_sentence(lemmatized_tokens)
```

```python
# This one is to keep all of the tokens in one list.
def all_lemmatize_sentence(tokens):
    lemmatizer = WordNetLemmatizer()
    lemmatized_sentence = []
    for i in range(len(tokens)):
        for word, tag in pos_tag(tokens[i]):
            if tag.startswith('NN'):
                pos = 'n'
            elif tag.startswith('VB'):
                pos = 'v'
            else:
                pos = 'a'
            lemmatized_sentence.append(lemmatizer.lemmatize(word, pos))
    return lemmatized_sentence
```

```python
# To create one big list and put every lemmatized token inside it
all_lemmatized_tokens = all_lemmatize_sentence(data_tokens)
```

```python
cleaned_tokens = []
for words in all_lemmatized_tokens:
    if words not in set(stopwords.words('english')):
        cleaned_tokens.append(words.lower())
    else:
        pass
```

## Determining word density

```python
# source : https://www.digitalocean.com/community/tutorials/how-to-perform-sentiment-analysis-in-python-3-using-the-natural-language-toolkit-nltk
def get_all_words(cleaned_tokens_list):
    for tokens in cleaned_tokens_list:
            yield tokens
```

```python
all_words = get_all_words(cleaned_tokens)
freq_dist = FreqDist(all_words)
```
